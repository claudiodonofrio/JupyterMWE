# Check FP files from ETC

```python
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
from halo import Halo 
import ipywidgets as widgets
from IPython.display import display, clear_output
from halo import Halo
```

```python
# Define data files
fp_data = ['DK-Sor_FFP2D_20240823.nc','DK-Sor_FFP2D_20240824.nc','DK-Sor_FFP2D_20240825.nc']
```

```python
# Provide some helper functions
```

```python
def select_fp(ds=''):
    if not ds: return []   
    ds = xr.open_dataset(ds)
    qf = ds.quality_flag.values
    ts = ds.time.values
    df = pd.DataFrame({'time': ts, 'quality_flag': qf})
    options = []
    for i in df.index:
        o = f'{df.time[i].strftime("%Y-%m-%d %X"), int(df.quality_flag[i])}'
        options.append((o, i))
    return options
```

```python
def plot_fp(ds='', fp=0):
    if not ds: return None    
    ds = xr.open_dataset(ds)
    return ds['FFP'][fp_widget.value,:,:].plot.contourf(levels=10)
```

```python
file_widget = widgets.Dropdown(
                options=fp_data,    
                description='File:',
                #value = None,
                disabled=False,
                )

fp_widget = widgets.Dropdown(
                options=select_fp(file_widget.value),    
                description='Footprint:',
                disabled=False,
                #"value = None,
                )

fp_plot = widgets.Output()
file_info = widgets.Output()

def file_change(change):
    fp_widget.options = select_fp(file_widget.value)    
    #ds = xr.open_dataset(file_widget.value)
    #with file_info:
    #    file_info.clear_output(wait=True)
    #    display(ds.info())



def fp_change(change):    
    ds = xr.open_dataset(file_widget.value)        
    with fp_plot:
        fp_plot.clear_output(wait=True)    
        fig, ax = plt.subplots()                
        ds['FFP'][fp_widget.value,:,:].plot.imshow(levels=10, ax=ax) 
        plt.show()
    
file_widget.observe(file_change)
fp_widget.observe(fp_change)
```

```python
display(widgets.HBox([file_widget, fp_widget]),widgets.VBox([fp_plot, file_info]))
```
