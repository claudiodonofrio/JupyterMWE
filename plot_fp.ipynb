```python
import xarray as xr
import ipywidgets as widgets
import pandas as pd
```

```python
dk1 = 'DK-Sor_FFP2D_20240823.nc'
dk1 = 'DK-Sor_FFP2D_20240824.nc'

ds = xr.open_dataset(dk1)
```

### Please select a timestep to plot, the second number is the quality flag
- 0 = good quality
- 3 = gap filled

```python
qf = ds.quality_flag.values
ts = ds.time.values
df = pd.DataFrame({'time': ts, 'quality_flag': qf})
options = []
for i in df.index:
    o = f'{df.time[i].strftime("%Y-%m-%d %X"), int(df.quality_flag[i])}'
    options.append((o, i))
```

```python
def plot_fp(time_index):
        ds['FFP'][time_index,:,:].plot.imshow(levels=10)
```

```python
widgets.interact(plot_fp, time_index=options)
```
